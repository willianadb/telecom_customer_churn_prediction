# Customer Churn Prediction Model for a Telecommunications Company

## Contextualization

The telecommunications company Interconnect wants to predict customer churn. If it is discovered that a user is planning to switch providers, the company will offer them promotional codes and special plan options. The marketing team at Interconnect has collected some personal data from its clientele, including information about their plans and contracts.

**Interconnect Services**

Interconnect primarily provides two types of services:

1. **`Landline Phone`**. The phone can be connected to multiple lines simultaneously.
2. **`Internet`**. The network can be established via a telephone line (DSL, digital subscriber line) or through a fiber optic cable.

Some other services provided by the company include:

- **Internet Security**: Includes antivirus software (DeviceProtection) and a malicious website blocker (OnlineSecurity).

- **Technical Support**: A dedicated technical support line (TechSupport).

- **Storage and Backup**: Cloud file storage and data backup (OnlineBackup).

- **Entertainment**: TV streaming (StreamingTV) and a movie directory (StreamingMovies).

Customers can choose between making a monthly payment or signing a 1 or 2-year contract. They can use various payment methods and receive an electronic invoice after the transaction.

**Data Description**

The data consists of files obtained from different sources:

- `contract.csv` — contract information;
- `personal.csv` — customer's personal data;
- `internet.csv` — information about internet services;
- `phone.csv` — information about phone services.

In each file, the `customerID` column contains a unique code assigned to each customer. The contract information is valid as of February 1, 2020.

Clarifications:

Target feature: the `'EndDate'` column is equal to `'No'`.

- Primary Metric: AUC-ROC.

- Additional Metric: Accuracy.

**Evaluation Criteria: Achieve an AUC-ROC ≥ 0.88**

AUC-ROC < 0.75 — 0 PPE

- 0.75 ≤ AUC-ROC < 0.81 — 4 PPE

- 0.81 ≤ AUC-ROC < 0.85 — 4.5 PPE

- 0.85 ≤ AUC-ROC < 0.87 — 5 PPE

- 0.87 ≤ AUC-ROC < 0.88 — 5.5 PPE

- AUC-ROC ≥ 0.88 — 6 PPE

> **Project Impact**: The ability to predict customer churn can significantly impact Interconnect's strategy. With the predictions generated by the model, the marketing team can identify at-risk customers and take proactive measures, such as offering personalized discounts, improving customer experience, and implementing loyalty programs. In this way, the company can increase retention, reduce costs associated with acquiring new customers, and consequently improve its profitability.

```python
# Enhanced function to convert column names to snake_case, preserving abbreviations like ID
def convert_to_snake_case(df):
    # Replace spaces with underscores and convert uppercase letters to snake_case, ignoring abbreviations like ID
    df.columns = [re.sub(r'(?<=[a-z])(?=[A-Z])', '_', col).replace(' ', '_').lower() for col in df.columns]
    return df
```

```python
# Function to identify outliers using the IQR
def identify_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers
```

### Implement Preprocessing Script

To make the data preprocessing replicable and organized, we developed a dedicated script for this step. The `preprocess_data.py` script handles tasks such as normalization, encoding categorical variables, handling missing values, and other necessary techniques to prepare the dataset for modeling.

Below is the code for the script, which you can reuse to preprocess new datasets:

#### Implement Script `preprocess_data.py`

```python
import os

# Path to the folder where the script will be saved (relative to the project directory)
script_path = os.path.join('..', '..', 'scripts', 'english', 'preprocess_data.py')

# Add debugging messages and check the directory
try:
    # Print the absolute path where the script will be saved
    print(f"Path to save the script: {os.path.abspath(script_path)}")
    
    # Check if the 'scripts/english' directory exists
    if not os.path.exists(os.path.dirname(script_path)):
        print(f"Directory {os.path.dirname(script_path)} does not exist. Creating...")
        os.makedirs(os.path.dirname(script_path), exist_ok=True)
    else:
        print(f"Directory {os.path.dirname(script_path)} already exists.")
    
    # Code for preprocess_data.py in English
    preprocess_data_code = """# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import os
import re

def convert_to_snake_case(df):
    # Replace spaces with underscores and convert uppercase letters to snake_case, ignoring abbreviations like ID
    df.columns = [re.sub(r'(?<=[a-z])(?=[A-Z])', '_', col).replace(' ', '_').lower() for col in df.columns]
    return df

def merge_data(files_list):
    try:
        # Load and convert column names to snake_case
        df_contract = convert_to_snake_case(pd.read_csv(files_list[0]))
        df_personal = convert_to_snake_case(pd.read_csv(files_list[1]))
        df_internet = convert_to_snake_case(pd.read_csv(files_list[2]))
        df_phone = convert_to_snake_case(pd.read_csv(files_list[3]))

        # Merge DataFrames
        df_merged = pd.merge(df_contract, df_personal, on='customer_id', how='left')
        df_merged = pd.merge(df_merged, df_internet, on='customer_id', how='left')
        df_merged = pd.merge(df_merged, df_phone, on='customer_id', how='left')

        return df_merged
    except Exception as e:
        print(f"Error merging files: {e}")
        raise

def preprocess_data(input_files, output_path):
    try:
        # Merge datasets
        df_merged = merge_data(input_files)

        # Convert the 'begin_date' and 'end_date' columns to datetime
        df_merged['begin_date'] = pd.to_datetime(df_merged['begin_date'], errors='coerce')
        df_merged['end_date'] = df_merged['end_date'].replace('No', pd.NaT)  # Replace 'No' with NaT
        df_merged['end_date'] = pd.to_datetime(df_merged['end_date'], errors='coerce')

        # Convert the 'total_charges' column to float, replacing empty spaces with NaN
        df_merged['total_charges'] = pd.to_numeric(df_merged['total_charges'], errors='coerce')

        # Handle missing values
        internet_columns = ['internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies']
        df_merged[internet_columns] = df_merged[internet_columns].fillna('No')
        df_merged['multiple_lines'] = df_merged['multiple_lines'].fillna('No')

        # Drop rows with NaN in 'total_charges'
        df_merged.dropna(subset=['total_charges'], inplace=True)

        # Create the target variable based on the 'end_date' column
        df_merged['target'] = df_merged['end_date'].notna().astype(int)

        # Contract duration feature
        df_merged['contract_duration'] = (df_merged['end_date'] - df_merged['begin_date']).dt.days
        df_merged['contract_duration'].fillna((pd.to_datetime('2020-02-01') - df_merged['begin_date']).dt.days, inplace=True)

        # Average charges
        df_merged['average_charges'] = df_merged['total_charges'] / df_merged['contract_duration']

        # Charges per SeniorCitizen
        df_merged['monthly_charges_per_senior_citizen'] = df_merged['monthly_charges'] * df_merged['senior_citizen']
        df_merged['total_charges_per_senior_citizen'] = df_merged['total_charges'] * df_merged['senior_citizen']

        # High monthly charges feature
        df_merged['high_monthly_charges'] = (df_merged['monthly_charges'] > df_merged['monthly_charges'].median()).astype(bool)

        # Long-term contract feature
        df_merged['long_term_contract'] = (df_merged['contract_duration'] > 365).astype(bool)

        # Adjust binary column types to 'bool'
        df_merged['high_monthly_charges'] = df_merged['high_monthly_charges'].astype(bool)
        df_merged['long_term_contract'] = df_merged['long_term_contract'].astype(bool)

        # List of categorical columns
        categorical_columns = ['type', 'paperless_billing', 'payment_method', 'gender', 'partner',
                               'dependents', 'internet_service', 'online_security', 'online_backup',
                               'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies',
                               'multiple_lines', 'senior_citizen']

        # Encode categorical variables using One-Hot Encoding
        df_encoded = pd.get_dummies(df_merged, columns=categorical_columns, drop_first=True)

        # List of numerical columns to normalize
        numeric_features = ['monthly_charges', 'total_charges', 'contract_duration', 'average_charges',
                            'monthly_charges_per_senior_citizen', 'total_charges_per_senior_citizen']

        # Apply scaler to numerical columns
        scaler = StandardScaler()
        df_encoded[numeric_features] = scaler.fit_transform(df_encoded[numeric_features])

        # Move the 'target' column to the last position
        cols = list(df_encoded.columns)
        cols.append(cols.pop(cols.index('target')))
        df_encoded = df_encoded[cols]

        # Rename 'senior_citizen_1' to 'senior_citizen_Yes' for better understanding
        if 'senior_citizen_1' in df_encoded.columns:
            df_encoded.rename(columns={'senior_citizen_1': 'senior_citizen_Yes'}, inplace=True)

        # Save the preprocessed DataFrame to the specified path
        os.makedirs(os.path.dirname(output_path), exist_ok=True)
        df_encoded.to_csv(output_path, index=False)
        print(f"Preprocessed dataset exported to: {output_path}")

    except Exception as e:
        print(f"Error during preprocessing: {e}")
        raise
"""

    # Save the code in 'preprocess_data.py'
    with open(script_path, 'w', encoding='utf-8') as file:
        file.write(preprocess_data_code)
    
    print(f"The script preprocess_data.py was successfully saved at {os.path.abspath(script_path)}!")

except Exception as e:
    print(f"Error saving the script: {e}")
```

```python
# Function to standardize formats (e.g., dates) and compare DataFrames
def standardize_and_compare(df1, df2, tolerance=1e-6):
    try:
        # Initial comparison to check for total equality of DataFrames
        comparison = df1.equals(df2)
        print(f"Are the DataFrames equal? {comparison}")

        # Check if the columns are the same (same name and order)
        columns_equal = df1.columns.equals(df2.columns)
        print(f"Are the columns equal? {columns_equal}")

        # Check if the data types are the same
        types_equal = df1.dtypes.equals(df2.dtypes)
        print(f"Are the data types equal? {types_equal}")

        # Align the data types
        df2_aligned = df2.astype(df1.dtypes)

        # Check for missing values
        missing_values_equal = df1.isna().sum().equals(df2_aligned.isna().sum())
        print(f"Is the number of missing values equal? {missing_values_equal}")

        # Check if the indices are the same
        indices_equal = df1.index.equals(df2_aligned.index)
        print(f"Are the indices equal? {indices_equal}")

        # Standardize 'end_date' to datetime
        df1['end_date'] = pd.to_datetime(df1['end_date'], errors='coerce')
        df2_aligned['end_date'] = pd.to_datetime(df2_aligned['end_date'], errors='coerce')

        # Compare specific columns, e.g., 'end_date'
        differences_end_date = compare_columns(df1, df2_aligned, 'end_date')
        if differences_end_date.any():
            print("Differences in the 'end_date' column after adjustment:")
            print(df1[differences_end_date][['end_date']].head())
            print(df2_aligned[differences_end_date][['end_date']].head())
        else:
            print("No significant differences found in the 'end_date' column.")

        # Compare numerical columns with tolerance
        numerical_columns = ['total_charges', 'total_charges_per_senior_citizen']
        for col in numerical_columns:
            differences_col = compare_columns(df1, df2_aligned, col, tolerance=tolerance)
            if differences_col.any():
                print(f"Differences in the '{col}' column (tolerance {tolerance}):")
                print(df1[differences_col][[col]].head())
                print(df2_aligned[differences_col][[col]].head())
            else:
                print(f"No significant differences found in the '{col}' column.")

    except Exception as e:
        print(f"Error during DataFrame comparison: {e}")
        raise
```

```python
# Function to compare columns
def compare_columns(df1, df2, column, tolerance=None):
    # Comparison for datetime columns
    if pd.api.types.is_datetime64_any_dtype(df1[column]):
        differences_col = ~df1[column].fillna(pd.Timestamp('1900-01-01')).eq(df2[column].fillna(pd.Timestamp('1900-01-01')))
    # Comparison for numerical columns
    elif tolerance is not None:
        differences_col = np.abs(df1[column] - df2[column]) > tolerance
    else:
        differences_col = df1[column] != df2[column]
    
    return differences_col
```

### Apply SMOTE to the Training Set

In this step, we apply the **SMOTE (Synthetic Minority Over-sampling Technique)** resampling technique to address class imbalance. SMOTE generates new synthetic samples of the minority class, balancing the proportion of samples between classes. This is crucial for improving model performance, especially in **recall** and **precision** metrics.

Applying SMOTE is particularly important when the minority class represents a small proportion of the data, as in our case, where customers who canceled represent approximately 26.6% of the total.

- **min_child_weight:** Minimum sum of the weights of all observations required in a child node. Explored values: [1, 2, 3, 4].
- **subsample:** Proportion of samples used to grow each tree. Explored values: [0.6, 0.8, 0.9, 1.0].
- **colsample_bytree:** Proportion of features used per tree. Explored values: [0.6, 0.8, 0.9, 1.0].

##### Adjust Probability Threshold

After optimizing the model's hyperparameters with `RandomizedSearchCV` and `GridSearchCV`, we identified that an additional adjustment in the probability threshold could improve performance concerning the recall of the "Churned" class. Therefore, we applied threshold adjustment using **Youden's J statistic** to find the optimal point.

```python
# Calculate the ROC Curve
fpr, tpr, thresholds = roc_curve(y_val, y_prob_xgb)
```

```python
# Calculate Youden's J Statistic to Find the Best Threshold
J = tpr - fpr
ix = argmax(J)
best_threshold = thresholds[ix]
print(f'Best Threshold = {best_threshold:.3f}')
```

**Threshold Adjustment Result**
- Best Threshold Found: 0.248

**Applying the New Threshold**

```python
# Apply the New Threshold to Make Predictions
y_pred_custom_threshold = (y_prob_xgb >= best_threshold).astype(int)
```

```python
# Evaluate the Model with the New Threshold
xgb_custom_auc_roc = roc_auc_score(y_val, y_pred_custom_threshold)
xgb_custom_accuracy = accuracy_score(y_val, y_pred_custom_threshold)
xgb_custom_class_report = classification_report(y_val, y_pred_custom_threshold, output_dict=True)
xgb_custom_conf_matrix = confusion_matrix(y_val, y_pred_custom_threshold)
```

```python
# Display the Results
print(f'AUC-ROC with adjusted threshold: {xgb_custom_auc_roc:.4f}')
print(f'Accuracy with adjusted threshold: {xgb_custom_accuracy:.4f}')
```

**Threshold Adjustment Result**
- **Best Threshold Found**: 0.248
- **AUC-ROC with Adjusted Threshold**: 0.8025
- **Accuracy with Adjusted Threshold**: 0.7801


##### Additional Visualization: ROC Curve

To enrich the analysis and demonstrate the impact of threshold adjustment, we compared the model with the default threshold (0.5) and the adjusted threshold (0.248).

```python
# Calculate the ROC Curve for the Adjusted Thresholds
fpr_default, tpr_default, _ = roc_curve(y_val, y_prob_xgb)
fpr_adjusted, tpr_adjusted, _ = roc_curve(y_val, y_pred_custom_threshold)
```

```python
# Plot the ROC Curve
plt.figure(figsize=(8, 6))
RocCurveDisplay(fpr=fpr_default, tpr=tpr_default).plot(label="Default Threshold (0.5)", ax=plt.gca())
RocCurveDisplay(fpr=fpr_adjusted, tpr=tpr_adjusted).plot(label=f"Adjusted Threshold ({best_threshold:.3f})", ax=plt.gca())
plt.title('ROC Curve - Threshold Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

**Justification for Threshold Adjustment**

The threshold adjustment was performed to increase the recall of the "Churned" class, which is crucial in the customer churn scenario. The goal here is to capture more customers prone to churn, even if this results in an increase in false positives.

However, we observed that by adjusting the threshold, the AUC-ROC decreased from 0.8971 to 0.8025, and accuracy also saw a slight drop from 0.8559 to 0.7801. This trade-off highlights the need to balance the model's sensitivity with its precision.

In customer churn scenarios, it is vital to find a balance between recall and precision. Increasing the recall of the "Churned" class allows the company to identify more at-risk customers, enabling preventive actions. However, higher recall may bring more false positives, raising operational costs. Therefore, we chose to assess the effectiveness of the adjusted threshold based on business objectives.

**Conclusion**

After the adjustments, the XGBoost model demonstrated a satisfactory balance between precision and recall with the default threshold of 0.5. Although the threshold adjustment improved the recall of the "Churned" class, the reduction in AUC-ROC and accuracy indicates that the model became more sensitive but less precise in distinguishing between customers who churn and those who do not.

We decided to maintain the default threshold, as it offers a better overall balance for this specific customer churn case. This balance is essential to provide more reliable predictions aligned with business objectives.

The XGBoost model not only shows good performance but also provides flexibility for future adjustments if business needs change. This makes it a robust choice for production applications, with the potential for continuous adaptations to enhance predictive capabilities.

##### Adjust Probability Threshold

After optimizing the model's hyperparameters with `RandomizedSearchCV` and `GridSearchCV`, we identified that an additional adjustment in the probability threshold could enhance the model's performance regarding the recall of the "Churned" class. To achieve this, we applied threshold adjustment using **Youden's J statistic** to find the optimal point.

**Threshold Adjustment**

```python
# Calculate the ROC Curve
fpr, tpr, thresholds = roc_curve(y_val, y_prob_lgbm)
```

```python
# Calculate Youden's J Statistic to Find the Best Threshold
J = tpr - fpr
ix = np.argmax(J)
best_threshold = thresholds[ix]
print(f'Best Threshold = {best_threshold:.3f}')
```

**Threshold Adjustment Result**
- Best Threshold Found: 0.248

**Applying the New Threshold**

```python
# Apply the New Threshold to Make Predictions
y_pred_custom_threshold = (y_prob_lgbm >= best_threshold).astype(int)
```

```python
# Evaluate the Model with the New Threshold
lgbm_custom_auc_roc = roc_auc_score(y_val, y_pred_custom_threshold)
lgbm_custom_accuracy = accuracy_score(y_val, y_pred_custom_threshold)
lgbm_custom_class_report = classification_report(y_val, y_pred_custom_threshold, output_dict=True)
lgbm_custom_conf_matrix = confusion_matrix(y_val, y_pred_custom_threshold)
```

```python
# Display the Results
print(f'AUC-ROC with adjusted threshold: {lgbm_custom_auc_roc:.4f}')
print(f'Accuracy with adjusted threshold: {lgbm_custom_accuracy:.4f}')
```

**Threshold Adjustment Result:**
- **Best Threshold Found**: 0.248
- **AUC-ROC with Adjusted Threshold**: 0.8085
- **Accuracy with Adjusted Threshold**: 0.7905

##### Additional Visualization: ROC Curve

To enrich the analysis and demonstrate the impact of the threshold adjustment, below is the ROC curve comparing the model with the default threshold (0.5) and the adjusted threshold (0.248).

```python
# Calculate the ROC Curve for the Adjusted Thresholds
fpr_default, tpr_default, _ = roc_curve(y_val, y_prob_lgbm)
fpr_adjusted, tpr_adjusted, _ = roc_curve(y_val, y_pred_custom_threshold)
```

```python
# Plot the ROC Curve
plt.figure(figsize=(8, 6))
RocCurveDisplay(fpr=fpr_default, tpr=tpr_default).plot(label="Default Threshold (0.5)", ax=plt.gca())
RocCurveDisplay(fpr=fpr_adjusted, tpr=tpr_adjusted).plot(label=f"Adjusted Threshold ({best_threshold:.3f})", ax=plt.gca())
plt.title('ROC Curve - Threshold Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

The ROC curve above demonstrates that by lowering the threshold, the model becomes more sensitive (increasing the true positive rate) but also generates more false positives. This change highlights the trade-off between recall and precision when adjusting the threshold.

**Justification for Threshold Adjustment**

The probability threshold adjustment was performed with the goal of increasing the recall of the "Churned" class, which is essential in the customer churn scenario. By adjusting the threshold to 0.248, the LightGBM model became more sensitive to detecting customers prone to churn. However, this adjustment resulted in a reduction of the AUC-ROC from 0.8806 to 0.8085 and a slight decrease in accuracy from 0.8445 to 0.7905. Despite this change, we prioritized recall, as it is more relevant in the context of predicting customers who are likely to cancel their services.

**Conclusion**

With the adjustments made, including the threshold adjustment, we conclude that the LightGBM model demonstrated solid performance with the default threshold of 0.5, offering a balance between precision and recall. Although the threshold adjustment improved the recall of the "Churned" class, it also negatively impacted the AUC-ROC and accuracy. Therefore, in this specific scenario, we decided to maintain the default threshold, as it provides a more suitable balance for the customer churn problem.

> "The LightGBM model, with its hyperparameter and threshold adjustments, proved to be a valuable tool for predicting customer churn. Its flexibility and accuracy can be leveraged in retention strategies, potentially reducing revenue losses related to churn."

   - The **F1-score** for the "Churned" class is **0.6627**, demonstrating a reasonable balance between precision and recall. However, the lower recall suggests that there is room for improvement in identifying at-risk customers.

3. **Did Not Churn Class**:
   - The precision and recall for the "Did Not Churn" class are **0.8640** and **0.9265**, respectively. These results indicate that the model is effective in correctly identifying customers who will not churn.
   - The high recall for this class suggests that the model leans more towards avoiding false negatives (customers who churn but are predicted as "Did Not Churn").

4. **Confusion Matrix**:
   - The model successfully identified **718 customers** who did not churn, with only **57 false positives** (customers predicted as "Churned" who did not churn).
   - For the "Churned" class, there were **167 true positives**, but also **113 false negatives** (customers who churned but were predicted as "Did Not Churn").
   - While the model is efficient at identifying customers who will not churn, the number of false negatives suggests a need for improvement in the model's sensitivity to the "Churned" class.

**Summary**

The XGBoost model demonstrated solid performance on the test set, particularly in identifying customers who will not churn. However, there is room to enhance the detection of at-risk customers, possibly by adjusting the threshold or improving the features.
